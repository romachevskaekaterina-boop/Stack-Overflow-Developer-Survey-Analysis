```python
import pandas as pd
```

```python
# завантаження та ознайомлення з файлами
pd.set_option('display.max_columns', None)
results=pd.read_csv("survey_results_public.csv",low_memory=False)
schema=pd.read_csv("survey_results_schema.csv")
results.head(5)
```

```python
# ЗАВДАННЯ 1: Загальна кількість респондентів
total_respondents=results["ResponseId"].nunique()
print(f"Усього респондентів, які взяли участь в опитуванні: {total_respondents} людей")
```

```python
# ЗАВДАННЯ 2: Аналіз повноти відповідей респондентів
questions=set(schema["qname"].unique())                             # виявляемо список питань з файлу схема
len_questions=len(questions)
common_questions=set(results.columns).intersection(set(questions))  # виявляемо спільні питання з обох файлів
len_common=len(common_questions)
print(f"Запитаннь в схемі: {len_questions}")
print(f"Запитань в таблиці відповідей: {len_common}")
```

```python
filtered_questions=results[list(common_questions)]                # виявляемо список питань з файлу результатів опитування
print(filtered_questions.shape)
```

```python
respondents_with_full_answer=filtered_questions.dropna()         # пририбаємо пустоти та рахуємо кількість повних відповідей
len_respondents=len(respondents_with_full_answer)
print(f"Кількість респондентів, які відповіли на всі запитання: {len_respondents}")
```

```python
# ЗАВДАННЯ 3: Статистичний аналіз досвіду респондентів
clean_workexp=results["WorkExp"].dropna()                       # прибираємо пусті значення
mean=clean_workexp.mean().round(0)                              # рахуємо показники
mode=clean_workexp.mode()[0]
median=clean_workexp.median()
print(f"Міри центральної тенденції досвіду роботи:")            # виводимо результати
print(f"Середнє: {mean} років")
print(f"Мода: {mode} років")
print(f"Медіана: {median} років")
```

```python
# ЗАВДАННЯ 4: Аналіз віддаленої роботи
format_of_work=results["RemoteWork"].unique()                  # виводимо список варіантів відповідей 
count_of_remote=len(results[results["RemoteWork"]=="Remote"])  # шукаємо та рахуємо кількість відаллено працюючих
print(f"Кількість респондентів, які працюють віддалено: {count_of_remote} людей")
```

```python
# ЗАВДАННЯ 5: Визначення популярності Python
results['worked_with_python'] = results['LanguageHaveWorkedWith'].str.contains('Python', case=False, na=False)
work_with_python = results['worked_with_python'].sum()  
popularity_of_python=round((work_with_python/total_respondents)*100,1)                            
print(f"Усього використовують Python: {work_with_python} респондентів")
print(f"Скільки респондентів використовують Рython: {popularity_of_python}%")
```

```python
# ЗАВДАННЯ 6: Аналіз шляхів навчання програмуванню
online_edication=len(results[results["LearnCode"].str.contains("Online Courses", na=False)])
print(f"Всього {online_edication} респондентів навчалися програмувати онлайн")
```

```python
# ЗАВДАННЯ 7: Географічний аналіз компенсації Python-розробників
# виявляємо тих, хто працює з Python
python_dev=results[results["LanguageHaveWorkedWith"].str.contains("Python", na=False)] 
# Групуємо попередні результати по країнам та обраховуємо показники
money_by_country=python_dev.groupby("Country")["ConvertedCompYearly"].agg(["size","count","mean", "median"]).dropna().round(0)
print(money_by_country.tail(30))
```

```python
# ЗАВДАННЯ 8: Аналіз освіти найбільш оплачуваних спеціалістів
clean_salary = results.dropna(subset=["ConvertedCompYearly"])                   # прибираємо пусті значення 
top_5=clean_salary.sort_values("ConvertedCompYearly", ascending=False).head(5)  # сортуємо результати
print(top_5[["ConvertedCompYearly", "EdLevel"]])                                # виводимо резульати розрахунків додаючи колонку з рівнем освіти
```

```python
# ЗАВДАННЯ 9: Аналіз популярності Python по віковим категоріям 
clean_data=results.dropna(subset=["Age"])                                                # прибираємо пусті значення
All_lang_by_group=clean_data.groupby("Age")["ResponseId"].count()                        # групуємо всі відповіді по віковим категоріям
Python_lang_by_group=python_dev.groupby("Age")["ResponseId"].count()                     # групуємо тих, хто використовує Python по віковим категоріям
popularity=((Python_lang_by_group/All_lang_by_group)*100).round(0).astype(str) + "%"     # розраховуємо популярність Python по віковим категоріям
print(f"Популярність Python за віковими категоріями:\n {popularity}")
```

```python
# ЗАВДАННЯ 10: Аналіз індустрій серед високооплачуваних віддалених працівників
# розраховуємо 75% перцентиль
above_salary=results["ConvertedCompYearly"].quantile(0.75)    
# фільтруємо таблицю за умовами
high_earners_remote=results[(results["ConvertedCompYearly"] > above_salary)&(results["RemoteWork"] == "Remote")]
# виявляємо в яких індустріях скільки респондентів працює
industry=high_earners_remote["Industry"].value_counts()
print(industry)
```
